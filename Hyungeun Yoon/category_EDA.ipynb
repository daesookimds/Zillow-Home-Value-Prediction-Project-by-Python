### merge 전 NaN 비율과 후 NaN 비율

```python
ratio_NaN_properties_2016 = cate_properties_2016.drop(["parcelid"], axis = 1).isnull().sum() / len(cate_properties_2016)
sort_NaN = ratio_NaN_properties_2016.sort_values(ascending = False)
ratio_of_NaN_2016 = pd.DataFrame(sort_NaN,columns = ["ratio"]).reset_index()

ratio_merge_df_2016 = merge_df_2016.drop(["parcelid"], axis = 1).isnull().sum() / len(merge_df_2016)
sort_merge_ratio = ratio_merge_df_2016.sort_values(ascending = False)
ratio_of_NaN_merge_2016 = pd.DataFrame(sort_merge_ratio,columns = ["ratio"]).reset_index()

pd.merge(ratio_of_NaN_2016, ratio_of_NaN_merge_2016, how = 'left', on='index', suffixes = ("_NaN", "_NaN_merge"))
```

```python
plt.figure(figsize = (10,5) )
plt.suptitle("ratio of NaN data")
plt.barh(sort_NaN.index, sort_NaN.values)
plt.show()
```

#### 제외 가능한 칼럼 : finishedsquarefeet15(같은 데이터가 존재하고 데이터의 양이 더 적음), finishedsquarefeet6(NaN값을 정확하게 채워줄 자료가 부족하고 calculatedfinishedsquarefeet 겹쳐짐), finishedfloor1squarefeet(같은 데이터가 존재, 단 32개의 데이터의 값은 다름), finishedsquarefeet13(NaN의 값을 0으로 생각할수도 있으나 데이터가 33개밖에 되지 않음), finishedsquarefeet12(대부분의 데이터가 calculatedfinishedsquarefeet 일치)

```python
columns_about_area = merge_df_2016[["parcelid", "finishedfloor1squarefeet", "calculatedfinishedsquarefeet", 'finishedsquarefeet6',\
                                          'finishedsquarefeet12', 'finishedsquarefeet13', 'finishedsquarefeet15', 'finishedsquarefeet50', 'lotsizesquarefeet']]
columns_about_area.tail()
```

```python
columns_about_area.describe()
```

#### calculatedfinishedsquarefeet(총면적)와 finishedsquarefeet15(총면적)를 비교해보면 같은 값이 3564개로 finishedsquarefeet15의 개수와 일치하므로 서로 같은 데이터를 가짐을 알 수 있고 따라서 NaN의 값은 단순히 누락된것을 예상할 수 있으나 서로 같은 데이터이므로 NaN값을 채워 활용할 필요가 없다. 따라서 더 많은 데이터를 포함하고 있는 calculatedfinishedsquarefeet를 사용하면 된다.

```python
area1 = columns_about_area[["calculatedfinishedsquarefeet", "finishedsquarefeet15"]]
area1[area1["calculatedfinishedsquarefeet"] == area1["finishedsquarefeet15"]].reset_index(drop=True).tail()
```

#### finishedsquarfeet6(완료되거나 완료되지 않은 공간면적)의 모든 데이터는 calculatedfinishedsquarefeet(총면적)의 값과 일치한다. finishedsquarefeet6는 unfinished and finished area라고 되어있지만 위의 결과에서 알 수 있듯이 모든 면적은 calculatedfinishedsquarefeet값과 일치하므로 unfinished area값을 알 수 있는 컬럼이 존재하지 않고 따라서 NaN값을 단순 누락으로 보고 calculatedfinishedsquarefeet의 값으로 채운다고 해도 서로 같은 데이터가되므로 사용할 필요가 없다 

```python
area2 =columns_about_area[["calculatedfinishedsquarefeet", "finishedsquarefeet6"]]
area2[area2["calculatedfinishedsquarefeet"] == area2["finishedsquarefeet6"]].reset_index(drop=True).tail()
```

#### finishedfloor1squarefeet(1층면적)보다 크거나 같은 calculatedfinishedsquarefeet(총면적)값이 6854개 인것으로 보아 2개의 값은 총면적이 1층면적보다 작다는 말이되고 잘못된 값으로 판단되어진다.

```python
area3 = columns_about_area[["calculatedfinishedsquarefeet", "finishedfloor1squarefeet"]]
area3[area3["calculatedfinishedsquarefeet"] == area3["finishedfloor1squarefeet"]].reset_index(drop=True).tail()
```

```python
area3[area3["calculatedfinishedsquarefeet"] > area3["finishedfloor1squarefeet"]].reset_index(drop=True).tail()
```

#### 총면적이 1층의 면적보다 작을수는 없기 때문에 calculatedfinishedsquarefeet의 값이 잘못되었고 finishedfloor1squarefeet의 값으로 대체 가능하다 

```python
area3[area3["calculatedfinishedsquarefeet"] < area3["finishedfloor1squarefeet"]].reset_index(drop=True).tail()
```

#### calculatedfinishedsquarefeet 의 값과 finishedsquarefeet12의 값이 일치하고 아래에서 조금은 변경이 가능하지만 큰 의미는 없을 듯...

```python
area4 =columns_about_area[["calculatedfinishedsquarefeet", "finishedsquarefeet12"]]
area4[area4["calculatedfinishedsquarefeet"] == area4["finishedsquarefeet12"]].reset_index(drop=True).tail()
```

#### calculatedfinishedsquarefeet(총면적)과 finishedsquarefeet13(주변생활공간)의 면적이 일치하는 개수가 33개로 finishedsquarefeet13의 개수와 일치한다. 따라서 같은 parcelid의  finishedsquarefeet12 값은 0으로 채울 수 있다.

```python
area6 =columns_about_area[["finishedfloor1squarefeet", "calculatedfinishedsquarefeet", "finishedsquarefeet50"]]
area6[area6["finishedfloor1squarefeet"] == area6["finishedsquarefeet50"] ].reset_index(drop=True).tail()
```

```python
area6[area6["finishedfloor1squarefeet"] < area6["finishedsquarefeet50"] ].reset_index()
```
