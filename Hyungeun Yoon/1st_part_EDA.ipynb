#### tax

- structuretaxvaluedollarcnt : The assessed value of the built structure on the parcel
- landtaxvaluedollarcnt : The assessed value of the land area of the parcel(토지세)
- taxvaluedollarcnt : The total tax assessed value of the parcel, structuretaxvaluedollarcnt + landtaxvaluedollarcnt
- taxamount : The total property tax(account the overall quality of the property, property values, square footage, home features and market conditions) assessed for that assessment year
- taxdelinquencyflag : 이전 해의 재산세
- taxdelinquencyyear : 미납된 재산세 납부시기

```python
tax_df = pd.DataFrame(mod_properties_2016, columns= ['parcelid', 'structuretaxvaluedollarcnt', 'landtaxvaluedollarcnt','taxvaluedollarcnt', 'taxamount'])

drop_tax_df = tax_df[tax_df['parcelid'].notnull() & tax_df['structuretaxvaluedollarcnt'].notnull() & tax_df['landtaxvaluedollarcnt'].notnull() & tax_df['taxvaluedollarcnt'].notnull() &tax_df['taxamount'].notnull()]
drop_tax_df
```

```python
cols = ['structuretaxvaluedollarcnt', 'landtaxvaluedollarcnt','taxvaluedollarcnt', 'taxamount']
sns.pairplot(drop_tax_df[cols])
plt.show()
```

#### 재산세 비율 계산

- 계산을 위해 NaN값을 제외

```python
for_ratio_tax = tax_df[["parcelid", "taxvaluedollarcnt", "taxamount"]].dropna(how='any')
drop_tax_df['ratio_tax'] = for_ratio_tax["taxamount"] / for_ratio_tax["taxvaluedollarcnt"]
drop_tax_df.tail(10)
```

```python
sns.pairplot(drop_tax_df.iloc[:,[1,2,-1]])
plt.plot()
```

### 면적에 따른 세금 비율

면적과 재산세 비율이 연관이 있을까 궁금해서 그래프를 그려보았지만 큰 상관은 없는듯... 

```python
squarefeet = properties_2016[["finishedsquarefeet15"]].dropna()
sqfeetandtax = pd.concat([for_ratio_tax[["parcelid", "ratio_tax"]] , squarefeet], join = 'inner', keys='parcelid', axis = 1)
sqfeetandtax
```

```python
figsize(12, 4)
plt.subplot(1, 2, 1)
plt.scatter(sqfeetandtax['a']['finishedsquarefeet15'], sqfeetandtax['p']['ratio_tax'])
plt.ylabel('ratio_tax')
plt.xlabel('finishedsquarefeet15')

plt.subplot(1, 2, 2)
plt.scatter(sqfeetandtax['a']['finishedsquarefeet15'], sqfeetandtax['p']['ratio_tax'])
plt.ylim(-0.2, 0.6)
plt.ylabel('ratio_tax')
plt.xlabel('finishedsquarefeet15')
plt.show()
```
